```python
def custom_matrix(shape, values=None):
    matrix = np.zeros(shape)  # Crea una matriz de ceros

    if values:
        for (i, j), val in values.items():
            i_adj, j_adj = i - 1, j - 1  # Ajustamos para que (1,1) sea la primera celda en lugar de (0,0)
            if 0 <= i_adj < shape[0] and 0 <= j_adj < shape[1]:  # Verifica que la posición es válida
                matrix[i_adj, j_adj] = val
            else:
                print(f"Advertencia: Posición ({i}, {j}) fuera de los límites.")

    return matrix

def markov(a, b, n):
    s = np.zeros(n, dtype=int)
    s[0] = 1
    for i in range(1, n):
        u = np.random.rand()
        if s[i - 1] == 0:
            s[i] = 1 if u <= a else 0
        elif s[i - 1] == 1:
            s[i] = 1 if u <= (1 - b) else 0
    return s
```

1) Compute filter vecor ($h_t$):

Notice that Given St we are in the classical Linear framework, where KF works for estimate the unobservable variables.

Then with Carter and Khon we can extract a draw from that series of unobservables.

```python
def Kalman_Filter(r, data, x_t, H, F, R, Q, St, mu_0, mu_1):
    T, n = data.shape
    mu = np.zeros((r, 1))

    filtered = np.zeros((T, r))
    filtered_err = np.zeros((T, r * r))
    Kalman_Gain = np.zeros((T, n * r))  # opcional
    h_t = np.zeros((r, 1))
    P_t = np.eye(r)

    for t in range(T):
        st = St[t]
        xt = x_t[t, 0]

        # 0) State-dependent mean
        mu[0, 0] = (1 - st) * mu_0 + st * mu_1 + st * xt

        # 1) Predict state and covariance
        h_t_hat = mu + F @ h_t
        P_t_hat = F @ P_t @ F.T + Q

        # ------ CONTROL DE MISSING DATA ------
        obs_mask = ~np.isnan(data[t, :])        # True donde hay dato
        yt = data[t, obs_mask].reshape(-1, 1)   # observaciones válidas

        H_mod = H[obs_mask, :]                  # filas de H correspondientes
        R_mod = R[np.ix_(obs_mask, obs_mask)]   # submatriz de R (filas y columnas)
        # -------------------------------------

        # Si no hay observaciones válidas, saltamos el update
        if yt.size == 0:
            h_t = h_t_hat
            P_t = P_t_hat
        else:
            # 2) Forecast error and its variance
            err = yt - H_mod @ h_t_hat
            v_err = H_mod @ P_t_hat @ H_mod.T + R_mod

            # 3) Kalman Gain
            K_t = P_t_hat @ H_mod.T @ np.linalg.inv(v_err)

            # 4) Update estimates
            h_t = h_t_hat + K_t @ err
            P_t = P_t_hat - K_t @ H_mod @ P_t_hat

        # 5) Store results
        filtered[t, :] = h_t.flatten()
        filtered_err[t, :] = P_t.flatten()

    return filtered, filtered_err

```

```python
def Carter_Kohn(r, filter, filter_err,x_t, F, Q, St, mu_0, mu_1):

    T = filter.shape[0]  # Obtener la dimensión de T
    # Inicialización
    ht_draw = np.zeros((T, r))
    # h_T|T
    mean1 = filter[T - 1, :]
    var1 = filter_err[T - 1, :].reshape(r, r)

    # Add a small diagonal constant to ensure positive definiteness
    var1 = var1 + np.eye(var1.shape[0]) * 1e-6
    h_t = np.random.multivariate_normal(mean1, var1)

    mu = np.zeros((r, 1))

    # Iteración hacia atrás
    for t in range(T - 1, -1, -1):

        # Asegurar que todo este en su sitio:
        st = St[t];
        xt = x_t[t,0];

        P_t_t = filter_err[t, :].reshape(r, r)

        # State dependent mean:
        # 0) State-dependent mean:
        mu[0, 0] = (1-st)*mu_0 + st*mu_1 + st*xt
        # Computar la ganancia
        H_star = F
        f_star = F @ P_t_t @ F.T + Q
        # Use pseudo-inverse instead of regular inverse # SOLUCION PROPUESTA POR GEMINI (Singular matrix) MUY BUENA, RESULEVE PROBELMA DE MATLAB TAMBIEN :)
        K_star = P_t_t @ H_star.T @ np.linalg.pinv(f_star)

        # h_t|t,h_t+1
        #mean2 = filter[t, :] + (K_star @ (h_t.T + (F @ filter[t, :].T)).T).T
        mean2 = filter[t, :] + K_star @ (h_t - mu.flatten() + F @ filter[t, :])
        # P_t|t,h_t+1
        var2 = P_t_t - K_star @ H_star @ P_t_t

        # Guardar la muestra
        ht_draw[t, :] = h_t

        # Tomar una nueva muestra de h_t
        var2 = var2 + np.eye(var2.shape[0]) * 1e-6
        h_t = np.random.multivariate_normal(mean2.flatten(), var2)

    return ht_draw
```

2) Compute states vecor ($S_t$):

Once we get our MS factor we are ready to use it to:

* **Hamilton Filter:** first filter probabilities of being in each state.
* **Khim & Nelson:** second to take a draw from that filter.



```python
def Hamilton_Filter(P, Y, x_t ,mu_0, mu_1, sigma20, sigma21):
    T = Y.shape[0]
    filtro = np.zeros((T, 2))

    # Paso (0): Valor inicial (xi_0|0)
    A = np.vstack((np.eye(2) - P, np.ones((1, 2))))
    E = np.array([0, 0, 1]).reshape(-1, 1)

    xi_pre = np.linalg.inv(A.T @ A) @ A.T @ E
    xi_pre = xi_pre.flatten()  # Convertir a vector

    for t in range(T):

        # Asegurar que todo este en su sitio:
        yt = Y[t];
        xt = x_t[t,0];
        # Paso (1): Forecast
        xi_for = np.array(xi_pre @ P)

        # Paso (2): Likelihood
        l0 = ((yt - mu_0) ** 2) / sigma20
        like0 = (1 / np.sqrt(2 * np.pi)) * (1 / np.sqrt(sigma20)) * np.exp(-l0 / 2)

        l1 = ((yt - (mu_1+xt)) ** 2) / sigma21
        like1 = (1 / np.sqrt(2 * np.pi)) * (1 / np.sqrt(sigma21)) * np.exp(-l1 / 2)

        # Paso (2b): Actualización usando la verosimilitud y xi_for
        joint_density = xi_for * np.array([like0 ,like1])
        xi_next = joint_density / np.sum(joint_density)

        # Paso (3): Guardar el filtro
        filtro[t, :] = xi_next

        # Paso (4): Preparar siguiente iteración
        xi_pre = xi_next

    return filtro
```

```python
def Kim_Nelson(P, filtro):
    T = filtro.shape[0]
    St = np.zeros(T, dtype=int)

    # Paso 1: último estado a partir del filtro (suavizado)
    smoothing_prob0 = filtro[T - 1, 0] / np.sum(filtro[T - 1, :])
    U = np.random.rand()

    if U >= smoothing_prob0:
        St[T - 1] = 1
    else:
        St[T - 1] = 0

    # Definimos las transiciones:
    p00 = P[0, 0]
    p01 = P[1, 0]
    p11 = P[1, 1]
    p10 = P[0, 1]

    # Paso 2: backward smoother
    for t in range(T - 2, -1, -1):  # desde T-2 hasta 0 (Python indexing)
        if St[t + 1] == 0:
            prob_p_state0 = p00 * filtro[t, 0]  # P(St=0 | St+1=0)
            prob_p_state1 = p10 * filtro[t, 1]  # P(St=1 | St+1=0)
        else:  # St[t+1] == 1
            prob_p_state0 = p01 * filtro[t, 0]  # P(St=0 | St+1=1)
            prob_p_state1 = p11 * filtro[t, 1]  # P(St=1 | St+1=1)

        total = prob_p_state0 + prob_p_state1
        if total == 0:
            prob_state0 = 0.5  # fallback en caso extremo
        else:
            prob_state0 = prob_p_state0 / total

        U = np.random.rand()
        St[t] = 0 if U < prob_state0 else 1

    return St
```

```python
def Kalman_Filter_xt(r, factor, sigma2_f, sigma2_v, St, mu_0, mu_1):


    T = factor.shape[0]  # número de observaciones
    filtered = np.zeros((T, r))
    filtered_err = np.zeros((T, r * r))

    x_t = np.array([[0]])  # estado inicial
    P_t = np.eye(r)         # covarianza inicial

    R = sigma2_f  # varianza del ruido de observación

    for t in range(T):
        # Término constante (drift) dependiente de St
        # Check if t is within the bounds of St
        if t < len(St):
            Az = (1 - St[t]) * mu_0 + St[t] * mu_1
        else:
            # Handle the case where t is out of bounds (e.g., use the last value of St)
            Az = (1 - St[-1]) * mu_0 + St[-1] * mu_1

        # Matriz de observación y transición (escalar en este caso)
        H = St[t] if t < len(St) else St[-1]
        F = St[t] if t < len(St) else St[-1]

        # Varianza del ruido del proceso
        Q = (1 - St[t])**2 * sigma2_v if t < len(St) else (1 - St[-1])**2 * sigma2_v



        # PREDICCIÓN
        x_t_hat = F * x_t
        P_t_hat = F**2 * P_t + Q

        # ERROR DE PREDICCIÓN
        err = factor[t] - H * x_t_hat - Az
        v_err = H**2 * P_t_hat + R

        # GANANCIA DE KALMAN
        K_t = (P_t_hat * H) / v_err

        # ACTUALIZACIÓN
        x_t = x_t_hat + K_t * err
        P_t = (1 - K_t * H) * P_t_hat

        # GUARDAR RESULTADOS
        filtered[t, :] = x_t.flatten()
        filtered_err[t, :] = P_t.flatten()

    return filtered, filtered_err
```

We are going to use 'common' priors (the conjugates ones we saw in class).

So in genereral (examples for AR(2)) we will have:

1) Set Priors:

* Normal Distributed Parameters:

$$
p(\phi) \sim N\left(\left(\begin{array}{c}
\phi_1^0 \\
\phi_2^0
\end{array}\right)_{\phi_0}, \left(\begin{array}{ccc}
\Sigma_{\phi_1} & 0 \\
 0 & \Sigma_{\phi_2}
\end{array}\right)_{\Sigma_\phi}\right)
$$

* Inverse Gamma Distributed Varainces:

$$
p(\sigma^2_\omega) \sim \Gamma^{-1}\left(\frac{T_{\omega_0}}{2}, \frac{\theta_{\omega_0}}{2}\right)
$$

$$
p(\sigma^2_\varepsilon) \sim \Gamma^{-1}\left(\frac{T_{\varepsilon_0}}{2}, \frac{\theta_{\varepsilon_0}}{2}\right)

* Dirichlet Distributed Transition Probabilities:

$$
p(p_{00}) \sim D(\alpha_{00}, \alpha_{01}) \text{ and } p(p_{11}) \sim D(\alpha_{11}, \alpha_{10})
$$



2) Set Conditional Posteriors:

* For Normal Distributed (mixed with Likelihood):

$$
H(\phi|\sigma^2_\omega, Y_t) \sim N(M^*, V^*)
$$
where:
$$
M^* = \left(\Sigma_{\phi_0}^{-1} + \frac{1}{\sigma_\omega^2}X'_tX_t\right)^{-1} \left(\Sigma_{\phi_0}^{-1}\phi_0 + \frac{1}{\sigma_\omega^2}X'_tY_t\right)
$$
$$
V^* = \left(\Sigma_{\phi_0}^{-1} + \frac{1}{\sigma_\omega^2}X'_tX_t\right)^{-1}
$$


* For Inverse Gamma Distributed (mixed with Likelihood):

$$
H(\sigma_\omega^2|\phi, Y_t) \sim \Gamma^{-1}\left(\frac{T_{\omega_1}}{2}, \frac{\theta_{\omega_1}}{2}\right)
$$

where:

$$
T_{\omega_1} = T_{\omega_0} + T
$$
$$
\theta_{\omega_1} = \theta_{\omega_0} + (Y_t - \phi X_t)'(Y_t - \phi X_t)
$$


* For Dirichlet Distributed (mixed with Likelihood):

$$
H(p_{00}|S) \sim D(\alpha_{00} + \eta_{00}, \alpha_{01} + \eta_{01})$$

$$
